```python
query = """
SELECT *
FROM balances;
"""

balances= pd.read_sql(query,con)
print(balances)
```

```python
def multiply_balance(balance, currency):
  if currency == 'CZK':
    balance = balance * 1
    return balance
  if currency == 'USD':
    balance = balance * 22
    return balance
  if currency == 'EUR':
    balance = balance * 25
    return balance
```

```python
import numpy as np
```

```python
balances['new_balance_CZK'] = np.vectorize(multiply_balance)(balances['balance'], balances['currency'])
```

```python
new_balances = balances.groupby(['client_id'],as_index=False).mean().drop('balance',axis=1)
```

```python
print(balances['client_id'].value_counts())
```
